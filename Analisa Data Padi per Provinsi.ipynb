### 2. Statistik tentang data
Berikut dilakukan analisis statistik deskriptif dari data padi per provinsi pada tahun 1993-2013

```python
import statistics, matplotlib, math
```

```python
s = data_padi.produksi.sum()
mi = data_padi.produksi.min()
ma = data_padi.produksi.max()
me = data_padi.produksi.mean()
md = data_padi.produksi.median()
print(s,mi,ma,me,md)
```

```python
mi1 = data_padi.produktivitas.min()
ma1 = data_padi.produktivitas.max()

print(mi1,ma1)
```

### 3. Visualisasi sederhana

Luas Panen Padi di Indonesia pada Tahun 1993-2013

```python
ax = data_padi.groupby('tahun').sum()['produksi']
ax = ax.plot(kind='line', title ="Jumlah Produksi per Tahun", legend=True, fontsize=12)
ax.set_xlabel("Tahun", fontsize=12)
ax.set_ylabel("Jumlah Produksi", fontsize=12)
ax
```

Luas Panen Padi di Indonesia pada Tahun 1993-2013

```python
ax1 = data_padi.groupby('tahun').mean()['luas_panen']
ax1 = ax1.plot(kind='line', title ="Luas Panen per Tahun", legend=True, fontsize=12)
ax1.set_xlabel("Tahun", fontsize=12)
ax1.set_ylabel("Luas Panen", fontsize=12)
ax1
```

```python
ax2 = data_padi.groupby('nama_provinsi').sum()['produksi']
ax2 = ax2.plot(kind='pie', title ="Jumlah Produksi per Provinsi", legend=True, fontsize=12)
ax2.set_xlabel("Provinsi", fontsize=12)
ax2.set_ylabel("Jumlah Produksi", fontsize=12)
ax2
```
